## Loading a graph from the collection
We import the `KGCOVID19` dataset from the `grape.datasets.kghub module`, and retrieve the graph by creating an instance of the `KGCOVID19` class. This knowledge graph has about 500K nodes and 18M edges.

```python
from grape.datasets.kghub import KGCOVID19
kgcovid19 = KGCOVID19()
```

The report generated by GRAPE provides a wealth of information about the knowledge graph, including metrics such as the number of nodes and edges, the degree centrality of nodes, the most common node and edge types, and topological oddities that may be present in the graph.

Topological oddities are sets of nodes in the graph that may be derived by errors during the generation of the edge list, and can bias the results of topology-based models. Some common types of topological oddities include singleton nodes, which are nodes that are disconnected from all other nodes, and cyclic edges, which are edges that form a cycle in the graph. By identifying these oddities, you can take steps to address them and improve the quality of your knowledge graph data.
